### mysql

```python
# mysql module
import mysql.connector

# connect_database( hostname , username , password )
def connect_database(host_name="host_name", user_name="user_name", user_password="user_password"):
    mydb = mysql.connector.connect(
        host = host_name,
        port = 3306,
        user = user_name,
        password = user_password,
        database = "WAY_BACK_HOME"
    )

    cursor = mydb.cursor()

    return mydb, cursor

def close_database(mydb, cursor):
    cursor.close()
    mydb.close()

# 
def excute_query(query):
    mydb, cursor = connect_database()
    try:
        cursor.execute(query);

        if query.strip().lower().startswith("select"):
            return cursor.fetchall()
        else :
            mydb.commit()
            print("success!!!")

    except mysql.connector.Error as err:
        print(err)

    finally:
        close_database(mydb=mydb,cursor=cursor)
```

### 매매가격지수 테이블 옮기기

```python
# HOUSE_PRICE_INDEX_APART

mydb, cursor = connect_database()
```

```python
# house_price_index_apart 에서 데이터 추출
cursor.execute("SELECT * FROM HOUSE_PRICE_INDEX_APART;")
rows = cursor.fetchall()
```

```python
# 새로 설계한 property_price_index 에 데이터 삽입
for row in rows:
    
    # date
    year_month = row[0]

    # 각 지역별로 분류
    regions = {
        "수도권": row[1],
        "서울": row[2],
        "대구": row[3]
    }

    for region_name, price_index in regions.items():
        
        if price_index is not None: # 가격이 none이 아닐때만 삽입

            # 지역 이름에 해당하는 region_id 조회
            select_query = f"select region_id from regions where region_name = '{region_name}'"
            # print(select_query)
            cursor.execute(select_query)
            result = cursor.fetchone()

            if (result):
                # region_id 
                region_id = result[0]

                # propperty_price_index 에 삽입
                cursor.execute(
                    "INSERT INTO property_price_index (region_id, date, price_index) VALUES (%s, %s, %s)",
                    (region_id, year_month, price_index)
                )
                print(f"{region_name}의 {year_month} 데이터가 정상적으로 삽입되었습니다.")
            else:
                print(f"{region_name}은 regions 테이블에 존재하지 않습니다")
    
mydb.commit()

cursor.close()
mydb.close()
```

```python
cursor.close()
mydb.close()
```

### 주택 공급량 데이터 옮기기

```python
# HOUSE_SUPPLY_COPY
mydb, cursor = connect_database()
```

```python
cursor.execute("SELECT * FROM HOUSE_SUPPLY_COPY;")
rows = cursor.fetchall()
```

```python
# HOUSE_SUPPLY_COPY -> housing_supply에 데이터 삽입
for row in rows:

    region_name = row[0]
    households_count = row[1]
    housing_count = row[2]
    penetration_rate = row[3]
    year = row[4]

    # region_name에 해당하는 region_id 조회
    cursor.execute("SELECT region_id FROM regions WHERE region_name = %s", (region_name,))
    result = cursor.fetchone()
    
    if result:
        region_id = result[0]
        # housing_supply에 데이터 삽입
        cursor.execute(
            "INSERT INTO housing_supply (region_id, date, households_count, housing_count, penetration_rate) VALUES (%s, %s, %s, %s, %s)",
            (region_id, year, households_count, housing_count, penetration_rate)
        )
        print(f"{region_name}의 {year} 데이터가 housing_supply에 삽입되었습니다.")
    else:
        print(f"{region_name}은(는) regions 테이블에 존재하지 않습니다.")

mydb.commit()

cursor.close()
mydb.close()
```

### 가게 대출 금리 데이터 옮기기

```python
# debt
mydb, cursor = connect_database()
```

```python
cursor.execute("SELECT * FROM debt limit 17;") # 같은 값이 여러번 들어가 있어서
rows = cursor.fetchall()
```

```python
# debt -> loan_interest 데이터 삽입
for row in rows:

    area = row[0]
    loan_values = row[1:]

    # region_name에 해당하는 region_id 조회
    cursor.execute("SELECT region_id FROM regions WHERE region_name = %s", (area,))
    result = cursor.fetchone()
    
    if result:

        region_id = result[0]
        # loan_interest에 데이터 삽입

        for year_index, loan_sum in enumerate(loan_values):

            year = f"{2006 + year_index}-00-00"

            if loan_sum is not None:
                print(f"{region_id, area}의 {year} 데이터가 loan_interest에 삽입되었습니다. (loan_sum: {loan_sum})")
                
                cursor.execute(
                    "INSERT INTO loan_interest (region_id, date, loan_sum) VALUES (%s, %s, %s)",
                    (region_id, year, loan_sum)
                )
                
        
    else:
        print(f"{region_name}은(는) regions 테이블에 존재하지 않습니다.")

mydb.commit()

cursor.close()
mydb.close()
```

### 지역별 미분양 현황

```python
# unsold_houses
mydb, cursor = connect_database()
```

```python
# unsold_houses 테이블에서 데이터 조회
cursor.execute("SELECT year_date, area, unsold_num FROM unsold_houses;") 
rows = cursor.fetchall()
```

```python
# unsold_houses -> unsold_status 데이터 삽입
for row in rows:

    year_date = row[0]
    area = row[1]
    unsold_num = row[2]

    # region_name에 해당하는 region_id 조회
    cursor.execute("SELECT region_id FROM regions WHERE region_name = %s", (area,))
    result = cursor.fetchone()
    
    if result:

        region_id = result[0]

        # unsold_status 데이터 삽입
        year = f"{year_date[:4]}-{year_date[4:6]}-00"
        # print(year)
        # break
        if unsold_num is not None:  # 미분양 수가 None이 아닐 경우에만 삽입
            cursor.execute(
                "INSERT INTO unsold_status (region_id, date, unsold_sum) VALUES (%s, %s, %s)",
                (region_id, year, unsold_num)
            )
            print(f"{area}의 {year} 데이터가 unsold_status에 삽입되었습니다. (unsold_num: {unsold_num})")
        
    else:
        print(f"{region_name}은(는) regions 테이블에 존재하지 않습니다.")

mydb.commit()

cursor.close()
mydb.close()
```
