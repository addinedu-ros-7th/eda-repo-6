```python
import mysql.connector
import pandas as pd

def connect_database(host_name="host_name", user_name="user_name", user_password="user_password"):
    mydb = mysql.connector.connect(
        host=host_name,
        port=3306,
        user=user_name,
        password=user_password,
        database="WAY_BACK_HOME"
    )
    cursor = mydb.cursor()
    return mydb, cursor

def close_database(mydb, cursor):
    cursor.close()
    mydb.close()

def excute_query(query, params=None):
    mydb, cursor = connect_database()
    try:
        if params:
            cursor.execute(query, params)  
        else:
            cursor.execute(query)
        if query.strip().lower().startswith("select"):
            return cursor.fetchall()
        else:
            mydb.commit()
            print("success!!!")
    except mysql.connector.Error as err:
        print(f"Error: {err}")
    finally:
        close_database(mydb=mydb, cursor=cursor)

debt_final = pd.DataFrame(debt_final)


def insert_data():
    for i, row in debt_final.iterrows():
        query = """
            INSERT INTO debt (area, 2006년, 2007년, 2008년, 2009년, 2010년, 2011년, 2012년, 2013년, 2014년, 2015년, 2016년, 2017년, 2018년, 2019년, 2020년, 2021년, 2022년, 2023년)
            VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
        """
        params = (
            row['지역코드'], row['2006년'], row['2007년'], row['2008년'], row['2009년'], row['2010년'],
            row['2011년'], row['2012년'], row['2013년'], row['2014년'], row['2015년'], row['2016년'],
            row['2017년'], row['2018년'], row['2019년'], row['2020년'], row['2021년'], row['2022년'], row['2023년']
        )
        excute_query(query, params)

insert_data()

```
